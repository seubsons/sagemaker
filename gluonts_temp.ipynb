```python
import xarray as xr
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from prophet import Prophet
from matplotlib.dates import date2num

from gluonts.mx.model.deepar import DeepAREstimator
from gluonts.mx.distribution import ZeroInflatedNegativeBinomialOutput, StudentTOutput #likelihood
from gluonts.mx.trainer.learning_rate_scheduler import LearningRateReduction
from gluonts.mx.trainer import Trainer
from gluonts.mx.trainer.model_averaging import ModelAveraging, SelectNBestSoftmax, SelectNBestMean

from gluonts.evaluation import Evaluator
from gluonts.evaluation.backtest import make_evaluation_predictions
from gluonts.dataset.common import ListDataset
from gluonts.dataset.util import to_pandas

latitudes = np.flipud(np.arange(65) * 0.25 + 5)
longitudes = np.arange(37) * 0.25 + 97
```

```python
ds = xr.open_dataset("data/temp.nc")
df = ds.to_dataframe()
df.reset_index(inplace=True)
df['date'] = df['date'].dt.date
df['lat'] = [latitudes[i] for i in df['lat']]
df['lon'] = [longitudes[i] for i in df['lon']]
df['lat_lon'] = df['lat'].astype(str) + ', ' + df['lon'].astype(str)
df = df.groupby(['date', 'lat_lon'])['temp'].max().reset_index()
pivoted_df = df.pivot(index=['date'], columns='lat_lon', values='temp').reset_index()
pivoted_df = pivoted_df.set_index('date')
pivoted_df
```

```python
training_list = []
for data in pivoted_df:
    training_list.append({"start": pivoted_df.index[0], "target": pivoted_df[data][:datetime.date(2023, 1, 1)].tolist()})
```

```python
training_data = ListDataset(
    training_list,
    freq="D"
)
```

```python
entry = next(iter(training_data))
train_series = to_pandas(entry)
```

```python
test_list = []
for data in pivoted_df:
    test_list.append({"start": pivoted_df.index[0], "target": pivoted_df[data][:datetime.date(2023, 11, 19)].tolist()})
```

```python
test_data = ListDataset(
    test_list,
    freq="D"
)
```

```python
entry = next(iter(test_data))
test_series = to_pandas(entry)
```

## Select the Model

```python
callbacks = [
    LearningRateReduction(objective="min",
                          patience=10,
                          base_lr=1e-3,
                          decay_factor=0.5,
                          ),
    ModelAveraging(avg_strategy=SelectNBestMean(num_models=2))
]


# estimator = DeepAREstimator(
#     freq="M",
#     prediction_length=306,
#     context_length=306,
#     num_layers = 5,
#     num_cells = 256,
#     distr_output=StudentTOutput(),
#     dropout_rate=0.01,
#     trainer=Trainer(#ctx = mx.context.gpu(),
#                     epochs=20,
#                     callbacks=callbacks))


# estimator = DeepAREstimator(
#     freq="1D",
#     prediction_length=7,  # Adjust as needed
#     trainer=Trainer(epochs=10)
# )


estimator = DeepAREstimator(
    freq="1D",
    prediction_length=306,
    context_length=306,
    num_layers = 2,
    num_cells = 20,
    distr_output=StudentTOutput(),
    dropout_rate=0.01,
    trainer=Trainer(#ctx = mx.context.gpu(),
                    epochs=5,
                    callbacks=callbacks))
```

```python
predictor = estimator.train(training_data)
```

```python
from gluonts import maybe
def plot(
    self,
    *,
    intervals=(0.5, 0.9),
    ax=None,
    color='g',
    name=None,
    show_label=True,
):

    ax = maybe.unwrap_or_else(ax, plt.gca)

    color = maybe.unwrap_or_else(
        color, lambda: ax._get_lines.get_next_color()
    )

    ax.plot(
        self.index.to_timestamp(),
        self.quantile(0.5),
        color=color,
        label=name,
    )
    

    for interval in intervals:
        if show_label:
            if name is not None:
                label = f"{name}: {interval}"
            else:
                label = interval
        else:
            label = None

        low = (1 - interval) / 2
        ax.fill_between(
            self.index.to_timestamp(),
            self.quantile(low),
            self.quantile(1 - low),
            alpha=0.5 - interval / 3,
            facecolor=color,
            label=label,
        )
```
